```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# On charge le fichier 
df = pd.read_csv('animes (6).csv')

# On affiche les 5 premières lignes
df.head()
```

```python
# On regarde combien de lignes on a avant
print("Animes avant nettoyage :", len(df))

# On ne garde que les animes où la note du pire épisode est supérieure à 0
df = df[df['Note_Pire_Ep'] > 0]

# On vérifie
print("Animes après nettoyage :", len(df))
```

```python
# 1. Calcul de l'écart (Est-ce que l'anime est constant ou en dents de scie ?)
df['Ecart'] = df['Note_Meilleur_Ep'] - df['Note_Pire_Ep']

# 2. Calcul du score de régularité (Plus c'est proche de 10, plus c'est stable)
df['Regularite'] = 10 - df['Ecart']

# 3. On affiche le résultat pour voir les nouvelles colonnes
df[['Anime', 'Note_Globale', 'Ecart', 'Regularite']].head()
```

```python
# On trie le tableau par 'Regularite' (du plus grand au plus petit)
top_regularite = df.sort_values(by='Regularite', ascending=False)

# On affiche le Top 5
top_regularite[['Anime', 'Regularite', 'Note_Globale']].head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- PARTIE 1 : CRÉATION DU SCORE ÉDITORIAL ---
# On s'assure que les colonnes de base sont là (au cas où)
df['Ecart'] = df['Note_Meilleur_Ep'] - df['Note_Pire_Ep']
df['Regularite'] = 10 - df['Ecart']

# La Formule Magique (Pondération)
# On donne plus d'importance à la Note Globale (60%) qu'à la Régularité (40%)
# car un chef-d'œuvre un peu irrégulier vaut mieux qu'un anime "moyen mais stable".
df['Score_Qualite'] = (df['Note_Globale'] * 0.6) + (df['Regularite'] * 0.4)

# --- PARTIE 2 : VISUALISATION (Le Graphique) ---
plt.figure(figsize=(12, 8)) # On définit une grande image

# On dessine le nuage de points
sns.scatterplot(
    data=df,
    x='Note_Globale',
    y='Score_Qualite',
    hue='Score_Qualite', # La couleur dépend du score
    size='Nb_Episodes',  # La taille du point dépend de la longueur de l'anime
    sizes=(50, 500),     # Taille min et max des bulles
    palette='viridis'    # Couleurs du bleu au jaune
)

# On ajoute la ligne rouge de référence
# Si un point est au-dessus : Notre score le valorise (Pépite cachée ou très stable)
# Si un point est en dessous : Notre score le pénalise (Trop irrégulier)
plt.plot([0, 10], [0, 10], color='red', linestyle='--', label='Ligne de référence')

plt.title("Matrice de Décision : Note Publique vs Qualité Éditoriale")
plt.xlabel("Note Globale (Avis des fans)")
plt.ylabel("Score Qualité (Calculé par l'algo)")
plt.legend()
plt.grid(True, alpha=0.3) # Une petite grille légère pour lire
plt.show()

# --- PARTIE 3 : LE CLASSEMENT FINAL ---
print("TOP 5 DES ANIMES À METTRE EN AVANT :")
top_final = df[['Anime', 'Score_Qualite', 'Note_Globale', 'Regularite']].sort_values(by='Score_Qualite', ascending=False)
display(top_final.head(5))

# --- PARTIE 4 : SAUVEGARDE ---
# On enregistre le fichier propre pour le livrable
df.to_csv('animes_final_score.csv', index=False)
print("Fichier 'animes_final_score.csv' sauvegardé avec succès !")
```

```python
# --- ETAPE DE RÉPARATION : On recalcule les scores ---

# 1. On nettoie (au cas où)
df = df[df['Note_Pire_Ep'] > 0]

# 2. On recrée les colonnes manquantes
df['Ecart'] = df['Note_Meilleur_Ep'] - df['Note_Pire_Ep']
df['Regularite'] = 10 - df['Ecart']
df['Score_Qualite'] = (df['Note_Globale'] * 0.6) + (df['Regularite'] * 0.4)

# 3. Maintenant que 'Score_Qualite' existe, on affiche un petit message
print("Colonnes recréées avec succès ! Tu peux lancer l'analyse.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. On regroupe par Studio et on fait la moyenne
top_studios = df.groupby('Studio')[['Score_Qualite', 'Note_Globale']].mean().sort_values(by='Score_Qualite', ascending=False)

# 2. Le Graphique
plt.figure(figsize=(12, 6))
sns.barplot(
    x=top_studios.index, 
    y=top_studios['Score_Qualite'], 
    palette='magma'
)

plt.title("Classement des Studios par Qualité Éditoriale")
plt.xticks(rotation=45, ha='right') # On penche le texte pour lire
plt.ylim(0, 10) # L'échelle de 0 à 10
plt.show()

print("Le studio N°1 est :", top_studios.index[0])
```
